## Permutation Tests

```python
# Instantiate the hypothesis report class
hp = HypothesisReport(sample1=before_yoga,
                      first_class = "before yoga",
                      sample2=after_yoga,
                      second_class="after yoga",
                      test_type="permutation",
                      alternative="two_sided",
                      test_statistic_type="signed_rank",
                      num_permutations=500_000)

# Build the hypothesis report
hp.build_output_file()
plt.close()
```

# Rank Sum Test -Trauma data

```python
non_trauma = [20.1, 22.9, 18.8, 20.9, 20.9, 22.7, 21.4, 20.0]
trauma = [38.5, 25.8, 22.0, 23.0, 37.6, 30.0, 24.5]

hp = HypothesisReport(sample1=trauma,
                      first_class = "trauma",
                      sample2=non_trauma,
                      second_class="non trauma",
                      test_type="permutation",
                      test_statistic_type="rank_sum",
                      alternative="two_sided",
                      test_value=0,
                      num_permutations=5_000_000)

hp.build_output_file()
plt.close()
```
