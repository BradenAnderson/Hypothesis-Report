# Test Drivers For the SampleAnalyzer Class

### Students_t with creativity data

```python
# test_type: Students_t, alternative: two_sided
analyzer = SampleAnalyzer(sample1=intrinsic,
                          sample2=extrinsic,
                          test_value=0,
                          alpha=0.05,
                          test_type="students_t",
                          alternative="two_sided")

sample_stats = analyzer.get_sample_statistics()

print("===================================================")
print("Data: Creativity")
print("Test Type: students_t")
print("Alternative: two_sided")
print(f"Agreeing Confidence Level: {analyzer.get_agreeing_conf_level()}\n")
print(f"Confidence Intervals: {analyzer.get_scipy_confidence_intervals()}\n")
print(f"Sample Statistics: {sample_stats}")
print(f"===================================================\n")

# test_type: Students_t, alternative: one_sided_right
analyzer = SampleAnalyzer(sample1=intrinsic,
                          sample2=extrinsic,
                          test_value=0,
                          alpha=0.05,
                          test_type="students_t",
                          alternative="one_sided_right")

sample_stats = analyzer.get_sample_statistics()

print("\n===================================================")
print("Data: Creativity")
print("Test Type: students_t")
print("Alternative: one_sided_right")
print(f"Agreeing Confidence Level: {analyzer.get_agreeing_conf_level()}\n")
print(f"Confidence Intervals: {analyzer.get_scipy_confidence_intervals()}\n")
print(f"Sample Statistics: {sample_stats}")
print(f"===================================================\n")

# test_type: Students_t, alternative: one_sided_left
analyzer = SampleAnalyzer(sample1=intrinsic,
                          sample2=extrinsic,
                          test_value=0,
                          alpha=0.05,
                          test_type="students_t",
                          alternative="one_sided_left")

sample_stats = analyzer.get_sample_statistics()

print("===================================================")
print("Data: Creativity")
print("Test Type: students_t")
print("Alternative: one_sided_left")
print(f"Agreeing Confidence Level: {analyzer.get_agreeing_conf_level()}\n")
print(f"Confidence Intervals: {analyzer.get_scipy_confidence_intervals()}\n")
print(f"Sample Statistics: {sample_stats}")
print(f"===================================================\n")
```
